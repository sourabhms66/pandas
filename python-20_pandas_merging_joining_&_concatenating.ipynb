### inner join: An inner join is a type of join operation in relational databases and data manipulation tools like pandas. When performing an inner join between two tables or DataFrames, only the rows with matching values in the specified key columns are included in the resulting DataFrame.

Here's an example to illustrate:

Consider two DataFrames df1 and df2:

df1:
   ID  Name
0   1   Bob
1   2  John
2   3  Mary

df2:
   ID  Age
0   1   30
1   2   25
2   4   35

If you perform an inner join on the 'ID' column, you'll get:

result:
   ID  Name  Age
0   1   Bob   30
1   2  John   25

In this result, only the rows with matching 'ID' values (1 and 2) from both DataFrames are included. Row with 'ID' 3 from df1 and row with 'ID' 4 from df2 are excluded because they don't have matching 'ID' values in both DataFrames.

So if we wanna join regs with students, for inner join, we're gonna use merge() method, pass how='inner', on='column_name_that_we_wanna_join' as arguments. This would return a merged dataframe the last number of students will be 25 as that's the maximum number of student we have in students dataframe even though we had larger student id in regs, as inner join only consider the commomns, it works lika an intersection of two sets

```python
students.merge(regs,how='inner',on='student_id')
```

### left join:

A left join, also known as a left outer join, is a type of join operation in relational databases and data manipulation tools like pandas. When performing a left join between two tables or DataFrames, all the rows from the left table (the first DataFrame) are included in the resulting DataFrame, along with matching rows from the right table (the second DataFrame). If there is no match for a row from the left table in the right table, null values are used to fill in the columns from the right table.

For example the code below produce a dataframe, if you look at the bottom, no students took numpy and c++ in regs but it's still printing with nan values, but if we have had performed left join these two rows wouldnt've printed

```python
courses.merge(regs,how='left',on='course_id')
```

### right join:

A right join, also known as a right outer join, is a type of join operation in relational databases and data manipulation tools like pandas. When performing a right join between two tables or DataFrames, all the rows from the right table (the second DataFrame) are included in the resulting DataFrame, along with matching rows from the left table (the first DataFrame). If there is no match for a row from the right table in the left table, null values are used to fill in the columns from the left table.

So, after running the code below the three student that we added above, their names won't be showed as they don't exit in regs and the last three student ids in regs dataframe they won't have any names as they don't exit in students, they will be replaced  by nan.

```python
students.merge(regs,how='right',on='student_id')
```

The nan values above will be exactly opposite if we have had done left join like the code below.

Similarly we can always reverse the dataframe order to achive the opposit join, like reverse of left join = right join

```python
#regs.merge(students,how='left',on='student_id') # reverse
students.merge(regs,how='left',on='student_id')
```

### outer join:

An outer join, also known as a full outer join, is a type of join operation in relational databases and data manipulation tools like pandas. When performing an outer join between two tables or DataFrames, all the rows from both tables are included in the resulting DataFrame. If there is no match for a row from one table in the other table, null values are used to fill in the columns from the missing table.

```python
students.merge(regs,how='outer',on='student_id')
```

### Question 1. find total revenue generated by the company

To solve this we'll join registration dataframe with courses dataframe with inner join, because we only need the prices for the courses in which the students are registerd.

```python
regs.merge(courses,how='inner',on='course_id')
```

Now we need only the ['price'] column, and this would return a series because one column in a dataframe is series

```python
regs.merge(courses,how='inner',on='course_id')['price']
```

Now we can apply sum on the series to get the total amount.

```python
regs.merge(courses,how='inner',on='course_id')['price'].sum()
```

### find out the orange cap holder of all the seasons (the batsman that made the highest run in a season)

```python
matches.head(2)
```

```python
delivery.head(2)
```

```python
temp_df = delivery.merge(matches,left_on='match_id',right_on='id')
```

Now if we groupBy using season and batsman and calculate the total runs made by each batsman in a season

```python
temp_df.groupby(['season','batsman'])['batsman_runs'].sum()
```

To convert to a dataframe

```python
temp_df.groupby(['season','batsman'])['batsman_runs'].sum().reset_index()
```

Now we'll sort the values, this is gonna give us a sorted series based on every season.

```python
temp_df.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False)
```

So there will be multiple batsman in a season, so here we'll use dropduplicates based on season column, drop the duplicate values and keep the fist one means the highest run in a season.

```python
temp_df.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False).drop_duplicates(subset=['season'],keep='first')
```

Now again if we sort the values using season

```python
temp_df.groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False).drop_duplicates(subset=['season'],keep='first').sort_values('season',ascending=False)
```
