### Question 5: Action movies with rating higher than 7.5

```python
movies.head(2)
```

Here we can't simply compare the two columns genres and imdb_rating because in ganres there are multiple ganres combined together with 'Action' and others, So first we have to find out the geners that have 'Action' in it and then compair them with rating.

To do that, we're gonna read the series as a string, this is gonna give us the whole thing as a string. Then to seperate every genre, we'll use split function to split the whole string based on'|' as this is what seperate the genres. Then we'll apply alogic to chech if 'Action' exits in the series, if does then we'll compare them with the ratings and this will give us the number of action movies that are more than 7.5 ratings

```python
movies['genres']
```

```python
movies['genres'].str
```

```python
movies['genres'].str.split('|')
```

```python
movies['genres'].str.split('|').apply(lambda x: 'Action' in x)
```

```python
movies[(movies['genres'].str.split('|').apply(lambda x: 'Action' in x)) & (movies['imdb_rating']==7.5)].shape[0]
#movies[(movies['genres'].str.contains('Action')) & (movies['imdb_rating']==7.5)].shape[0] # without using apply
```

### Question 6: write a function that can return the track record of 2 teams against each other --task

```python
ipl.head(2)
```

```python
team1 = ipl[ipl['WinningTeam']== 'Gujarat Titans'].shape[0]
team1
```

```python
team2 = ipl[ipl['WinningTeam']== 'Rajasthan Royals'].shape[0]
team2
```

```python
def track(team1,team2):
    t1 = ipl[ipl['WinningTeam']== team1].shape[0]
    t2 = ipl[ipl['WinningTeam']== team2].shape[0]
    return team1,':',t1,'   ',team2,':',t2
te1 = input("Enter Team 1: ")
te2 = input("Enter Team 2: ")

track(te1,te2)
```

### Add a column from a exiting column

Let's add a new column called 'Lead actor' and the value of the column will be the first name of the actors column

Let's take a look at the acrots name, the code below return a series of actors

```python
movies['actors']
```

As this is a series we'll take the string format of the series and split them on the basis of '|', this would return a series of list of the actors names

```python
movies['actors'].str.split('|')
```

If we try to fletch the first item of the series, it wouldn't work as it's gonna only return the first list of the series, that contains all the names of the actors

```python
movies['actors'].str.split('|')[0]
```

So we'll apply a logic to extract the first element of all the values of the column 'actors'

```python
movies['actors'].str.split('|').apply(lambda x: x[0])
```

The error we got because of the missing values, so first take care of the missing values eusing dropna() then apply the logic

```python
movies.info()
```

If we look at the RangeIndex the actors Non-Null values is 1624, the highest Non-null value is 1629 so 5 missing values.

Let's take care of them

```python
movies['actors'].dropna()
```

Now apply the same logic again, we are able to seperate the first actor name from the actors column

```python
movies['actors'].dropna().str.split('|').apply(lambda x:x[0])
```

Now add this sereis in a new column

```python
movies['Lead actor'] = movies['actors'].dropna().str.split('|').apply(lambda x:x[0])
```

Now if we check we can see the new column has been added

```python
movies.head()
```

If we check the shape it's 1629 that means the rows that had missing values those rows are still empty in the new column

# Important DataFrame Functions

### astype()

Use to change a data type

```python
movies.info()
```

So if we look at the basic information some of the datatype are int64 which is use to store very large number of integer value, we don't need that because if we check for all the cases the maximum value is not that big, so keep their data type as int64 is waste of memory, so here we can assign a data type that consume less memory to same the memory consumption by the dataset. Write now the memory usage is about 255kb 

```python
movies['is_adult'].max()
```

```python
movies['year_of_release'].max()
```

```python
movies['imdb_votes'].max()
```

So let's assign them a datatype that takes less memory by using astype()

```python
movies['is_adult'].astype('Int16')
```

To make the permanent change

```python
movies['is_adult'] = movies['is_adult'].astype('Int16')
```

Now let's do the same thing for the others

```python
movies['year_of_release'] = movies['year_of_release'].astype('Int16')
movies['imdb_votes'] = movies['imdb_votes'].astype('Int32')
```

Now if we see the memory consumption

```python
movies.info()
```

```python
ipl.info()
```

```python
ipl['ID'].max()
```

```python
ipl['ID'] = ipl['ID'].astype('Int32')
```

Similarly the columns 'team1','team2','Season' have repeated string values so we can change their datatype to 'category'

```python
ipl['Season'] = ipl['Season'].astype('category')
ipl['Team1'] = ipl['Team1'].astype('category')
ipl['Team2'] = ipl['Team2'].astype('category')
ipl['TossWinner'] = ipl['TossWinner'].astype('category')
```
