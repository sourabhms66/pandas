# GroupBy Objects

In Pandas, a groupby object is an intermediate data structure that allows you to perform grouping operations on a DataFrame. When you apply the groupby() function to a DataFrame, it splits the DataFrame into groups based on one or more specified keys and returns a groupby object.

This groupby object is essentially a mapping of keys to the corresponding group of rows from the original DataFrame. It doesn't actually perform any computation until you apply an aggregation function or another operation to it.

```python
movies.groupby('Genre')
```

```python
genre = movies.groupby('Genre')
```

It's a dataframe

```python
type(genre)
```

You can think of the groupby object as a way to efficiently handle grouped data without actually creating separate DataFrames for each group. It provides a convenient interface for performing operations on each group or aggregating data within each group.

Once you have a groupby object, you can apply various aggregation functions (such as sum, mean, count, etc.) to calculate summary statistics for each group, or you can apply custom functions using the apply() method.

```python
genre.sum()
```

```python
genre.min()
```

```python
genre.max()
```

# agg() aggregation method

In Pandas, the agg() method is used with groupby objects to perform aggregation operations on groups of data. It allows you to apply multiple aggregation functions simultaneously to the columns of the grouped DataFrame, producing a single result for each group.

The agg() method accepts one or more aggregation functions as arguments and applies them to the groups independently. These aggregation functions can be built-in functions such as 'mean', 'sum', 'min', 'max', or custom functions defined by the user.

```python
genre.sum()
```

We if apply built in function such as sum, min, max we can't apply them together, we have to do one by one like the code above but using agg() method we can apply multiple function at a time we'll pass a dictionary as an argument to this method, value -> columns, keys -> functions that we wanna apply on them, keys can be a list if we wanna apply multiple functions on one column.

```python
genre.agg({
    'Runtime': 'mean',
    'IMDB_Rating': ['mean','max','min'],
    'No_of_Votes': ['sum','max'],
    'Gross': 'mean',
    'Metascore':['mean','max']
})
```

We can also pass a list directly to calculate all the operation together like this, but to do that first we need to drop the columns that doesn't have only numerical values.

```python
genre1 = movies.drop(columns=['Released_Year','Series_Title','Director','Star1']).groupby('Genre') # dropping the columns that doesn't have only numerical values.
```

```python
genre1.agg(['max','mean','min','sum'])
```

### split (apply) combine

"Split-Apply-Combine" is a strategy often used in data analysis and manipulation, particularly in the context of programming languages like Python or R. It's a methodology for breaking down a problem into three steps:

Split: The data is divided into groups based on some criteria.
Apply: A function or operation is applied to each group independently.
Combine: The results from each group are combined into a final output.
This approach is especially useful for tasks such as aggregation, summarization, or transformation of data within each group. It allows for efficient and structured handling of large datasets, as operations are performed on smaller, more manageable subsets.

```python
genre.apply(min)
```

### Question: find ranking of each movie in the group according to IMDB score

```python
def rank_movies(group):
    group['rank']=group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genre.apply(rank_movies)
```

### Question: find normalized IMDB rating group wise



```python
def normalize(group):
    group['Normalized'] = (group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return group
```

```python
genre.apply(normalize)
```

### Create a function that can return the highest score of any batsman

```python
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('DA Warner')
```
