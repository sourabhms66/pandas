```python
class Fraction:
    # parameterized constructor
    def __init__(self,x,y):
        self.num = x
        self.den = y

    def __str__(self):
        return '{}/{}'.format(self.num,self.den)

    def __add__(self,other):
        new_num = (self.num*other.den)+(other.num*self.den)
        new_den = other.den*self.den
        return '{}/{}'.format(new_num,new_den)
    
    def decimal(self):
        return self.num/self.den

        
```

```python
import math
class Point:
    def __init__(self,x,y):
        self.x_cod = x
        self.y_cod = y
    
    def __str__(self):
        return '({},{})'.format(self.x_cod,self.y_cod)
    
    def uclidean_distance(self,other):
        return math.sqrt((self.x_cod-other.x_cod)**2+(self.y_cod-other.y_cod)**2)
    
    def distance_from_origin(self):
        #return math.sqrt(self.x_cod**2+self.y_cod**2)
        return self.uclidean_distance(Point(0,0))
    
class Line:
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c

    def __str__(self):
        return '{}x+{}y+{}=0'.format(self.a,self.b,self.c)
    
    def point_on_line(line,point):
        if line.a*point.x_cod+line.b*point.y_cod+line.c == 0:
            return 'The point is located on the line'
        else:
            return 'The point is not located on the line'
        
    def shortest_distance(line,point):
        return abs(line.a*point.x_cod+line.b*point.y_cod+line.c)/math.sqrt(line.a**2+line.b**2)
```

## The diamond problem:

The class that comes first, its method will get called

```python
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("buying a phone")

class Product:
    def buy(self):
        print("Product customer review")

class SmartPhone(Product,Phone): # method resolution order MRO
    pass

s = SmartPhone(800,'Apple',15)
s.buy()
```

```python
class Geometry:

    def area(self,radius):
        print("Area of the circle",3.14*radius*radius)
    
    def area(self,length,hight):
        print("Area of a rectangle",length*hight)

cal_area = Geometry()
cal_area.area(5)
```

```python
class Geometry:

    def area(self,r,l=0):
        if l ==0:
            print("Area of the circle",3.14*r*r)
        else:
            print("Area of the ractangle",r*l)
    

cal_area = Geometry()
cal_area.area(5,4)
```
